```python

import networkx as nx

# https://stackoverflow.com/questions/32935510/networkx-digraph-create-subgraph-digraph-by-node

def create_subgraph(g,sub_g,start_node):
    for n in g.predecessors(start_node):
        sub_g.add_edge(n,start_node)
        create_subgraph(g,sub_g,n)
```

```python
%matplotlib inline
G = nx.DiGraph()

G.add_nodes_from(["table1"])
G.add_nodes_from([2,3,4,5])
G.add_nodes_from([6,7,8,9,10])
G.add_edge("table1",3, process={"sql": "select * from TEMP"})
G.add_edge(3,2)
G.add_edge(3,5)
G.add_edge(4,5)
G.add_edge("table1",8)

G.add_edge(6,8)
G.add_edge(7,8)
G.add_edge(8,10)
G.add_edge(9,10)

G.add_edge(10,2)
A = nx.nx_agraph.to_agraph(G)
A.layout()
A.draw('graph.png')
```

```python
subG = nx.DiGraph()
create_subgraph(G, subG, 5)

B = nx.nx_agraph.to_agraph(subG)
B.layout()
B.draw('sub_graph.png')


```

```python
import numpy as np
import matplotlib.pyplot as plt
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2  # 0 to 15 point radii
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()

X = np.linspace(-np.pi, np.pi, 256,endpoint=True)
C,S = np.cos(X), np.sin(X)

plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-")
plt.plot(X, S, color="red", linewidth=2.5, linestyle="-")

plt.xlim(X.min()*1.1, X.max()*1.1)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
[r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

plt.ylim(C.min()*1.1,C.max()*1.1)
plt.yticks([-1, 0, +1],
[r'$-1$', r'$0$', r'$+1$'])

plt.show()
```
